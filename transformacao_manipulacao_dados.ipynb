# **Transformação e manipulação de dados**

## Preaparando o ambiente

```python
# Importando Pandas
import pandas as pd

# Importando NumPy
import numpy as np
```

```python
# Lendo o dataset
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
# Preparando o dataset
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

## Dados numéricos

```python
# Obtendo uma lista com os nomes das colunas
colunas = list(dados.columns)
colunas
```

```python
# Desagrupando dados com explode
dados = dados.explode(colunas[3:])
dados.head(3)
```

```python
# Resetando o index dos dados
dados.reset_index(inplace=True, drop=True)
dados.head(3)
```

```python
# Obtendo os tipos de dados
dados.info()

# object == varios tipos de dados em uma mesma coluna, tipo não identificado ou tipo string
```

```python
# Alterando o tipo de dados
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
# Alterando o tipo de dados
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
# Alterando o tipo de dados
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
# Removendo os caracteres especiais dos dados de preço
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
# Removendo caracteres especiais de mais de uma coluna
col_taxas = ['taxa_deposito', 'taxa_limpeza']
dados[col_taxas] = dados[col_taxas].map(lambda x: x.replace('$', '').replace(',', '').strip())

# apply == usado para uma series (linhas e colunas)
# map == usado para um dataframe (mais de uma series)
```

```python
# Alterando o tipo dos dados
dados[col_taxas] = dados[col_taxas].astype(np.float64)
dados[col_taxas].head(3)
```

```python
dados.info()
```

## Dados Textuais

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados.head(2)
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_local'].head(2)

# Regex:
# ^ == nega os valores a seguintes
```

```python
# Substituindo hífens por espaços em branco, a não ser em caso de palavras compostas
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)

''' O regex (?<!\w)-(?!\w) procura por hífens (-) que não estejam ligados à palvras, substituindo-os por ' '
Exemplo == "hello-world-this-is-a-test-hyphenated-word" == "hello world this is a test hyphenated-word"'''
```

```python
# Separando as palavras
dados['descricao_local'] = dados['descricao_local'].str.split()
dados['descricao_local'].head(3)
```

```python
# Removendo as {} e as '' de 'Comodidades'
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
dados['comodidades'].head(3)
```

```python
# Separando as palavras por vírgula
dados['comodidades'] = dados['comodidades'].str.split(',')
dados['comodidades'].head(3)
```

```python
# Verificando os dados de 'descricao_vizinhanca'
dados['descricao_vizinhanca'].head(3)
```

```python
# Convertendo as letras para tamanho minúsculo
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca'].head(3)
```

```python
# Usando regex para manter apenas caracteres alfanuméricos, '-' e '''
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'].head(3)
```

```python
# Substituindo hífens por espaços em branco, a não ser em caso de palavras compostas
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
dados['descricao_vizinhanca'].head(3)
```

```python
# Separando as palavras
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados['descricao_vizinhanca'].head(3)
```

## Dados de Tempo

```python
# Lendo a base de dados
dt_data = pd.read_json('/content/moveis_disponiveis.json')
dt_data.head(3)
```

```python
dt_data.info()
```

```python
# Convertendo os valores da coluna 'data' o tipo datetime
dt_data['data'] = pd.to_datetime(dt_data['data'])
```

```python
dt_data.info()
```

```python
dt_data.head(3)
```

```python
dt_data['data'].dt.strftime('%Y-%m').head(3)
```

```python
# Agrupando os dados por mês e número de vaga disponível
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```

```python
# Preenchendo os valores vazios com a string '0.0'
dt_data['preco'] = dt_data['preco'].fillna('0.0')
dt_data['preco'].head(3)
```

```python
# Removendo '$' e ','
dt_data['preco'] = dt_data['preco'].apply(lambda x: x.replace('$', '').replace(',', ''))
dt_data['preco'].tail(3)
```

```python
# Convertendo o tipo dos valores
dt_data['preco'] = dt_data['preco'].astype(np.float64)
dt_data['preco'].tail(3)
```
