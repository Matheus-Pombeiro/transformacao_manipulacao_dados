# **Transformação e manipulação de dados**

## Preaparando o ambiente

```python
# Importando Pandas
import pandas as pd
```

```python
# Lendo o dataset
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
# Preparando o dataset
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

## Dados numéricos

```python
# Obtendo uma lista com os nomes das colunas
colunas = list(dados.columns)
colunas
```

```python
# Desagrupando dados com explode
dados = dados.explode(colunas[3:])
dados.head(3)
```

```python
# Resetando o index dos dados
dados.reset_index(inplace=True, drop=True)
dados.head(3)
```

```python
# Obtendo os tipos de dados
dados.info()

# object == varios tipos de dados em uma mesma coluna, tipo não identificado ou tipo string
```

```python
# Importando NumPy
import numpy as np
```

```python
# Alterando o tipo de dados
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
# Alterando o tipo de dados
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
# Alterando o tipo de dados
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
# Removendo os caracteres especiais dos dados de preço
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
# Removendo caracteres especiais de mais de uma coluna
col_taxas = ['taxa_deposito', 'taxa_limpeza']
dados[col_taxas] = dados[col_taxas].map(lambda x: x.replace('$', '').replace(',', '').strip())

# apply == usado para uma series (linhas e colunas)
# map == usado para um dataframe (mais de uma series)
```

```python
# Alterando o tipo dos dados
dados[col_taxas] = dados[col_taxas].astype(np.float64)
dados[col_taxas].head(3)
```
